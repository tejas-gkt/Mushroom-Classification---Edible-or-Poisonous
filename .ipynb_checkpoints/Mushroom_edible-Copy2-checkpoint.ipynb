```python
# Splitting the data into features and target
x = data.drop('class', axis=1)
y = data['class']
```

```python
# Split into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42,stratify=y)
```

```python
# Manual encoding of Target feature: Edible = 1, Poisonous = 0
y_train = y_train.apply(lambda x: 1 if x=='e' else 0)
y_test = y_test.apply(lambda x: 1 if x=='e' else 0)
```

```python
training_results = pd.DataFrame()
training_results.index=['accuracy', 'precision', 'recall', 'f1', 'roc_auc']
testing_results = training_results.copy()
```

```python
# List of encoders to use
encoders = {
    'Binary Encoding': ce.BinaryEncoder(cols=['cap-shape', 'cap-surface', 'cap-color', 'does-bruise-or-bleed', 'gill-attachment', 'gill-color', 'stem-color', 'has-ring', 'ring-type', 'habitat', 'season'])
    #'Frequency Encoding': ce.CountEncoder(cols=['cap-shape', 'cap-surface', 'cap-color', 'does-bruise-or-bleed', 'gill-attachment', 'gill-color', 'stem-color', 'has-ring', 'ring-type', 'habitat', 'season'])
}

# Dictionary of classifiers
classifiers = {
    'LogisticRegression': LogisticRegression(random_state=42),
    'KNN': KNeighborsClassifier(n_neighbors=3),
    'DecisionTree': DecisionTreeClassifier(random_state=42),
    'RandomForest': RandomForestClassifier(n_estimators=100, random_state=42),
    'NaiveBayes': GaussianNB(),  # Naive Bayes
    'SVM': SVC(probability=True, random_state=42),  # SVM with probability=True for ROC-AUC
    'XGBClassifier': XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
    
}

# Define scoring metrics for cross-validation
scoring = ['accuracy', 'precision', 'recall', 'f1', 'roc_auc']



# Loop over both encoders
for encoder_name, encoder in encoders.items():
    print(f"\nUsing {encoder_name}:\n")
    
    # Loop over classifiers
    for clf_name, clf in classifiers.items():
        # Create a pipeline for each combination of encoder and classifier
        pipeline = Pipeline(steps=[
            ('encoder', encoder),  # Use the selected encoder
            ('scaler', StandardScaler()),  # Standardize features
            (clf_name, clf)  # Use the selected classifier
        ])
        
        # Perform cross-validation
        cv_results = cross_validate(
            pipeline, 
            x_train, 
            y_train, 
            cv=5,  # 5-fold cross-validation
            scoring=scoring,
            return_train_score=False
        )
        
        # Print metrics for each classifier and encoder
        print(f"\nClassifier: {clf_name} with {encoder_name}")
        print(f"Mean CV Accuracy: {cv_results['test_accuracy'].mean():.2f}")
        print(f"Mean CV Precision: {cv_results['test_precision'].mean():.2f}")
        print(f"Mean CV Recall: {cv_results['test_recall'].mean():.2f}")
        print(f"Mean CV F1-Score: {cv_results['test_f1'].mean():.2f}")
        if 'test_roc_auc' in cv_results:
            print(f"Mean CV ROC-AUC: {cv_results['test_roc_auc'].mean():.2f}")
        else:
            print("Mean CV ROC-AUC: N/A")
            
        training_results[f'{clf_name}_{encoder_name}'] = [cv_results['test_accuracy'].mean(), cv_results['test_precision'].mean(), cv_results['test_recall'].mean(), cv_results['test_f1'].mean(),cv_results['test_roc_auc'].mean() ]
            
            
        # Fit the model on the entire training data
        pipeline.fit(x_train, y_train)
        
        # Predict on the test set
        y_pred = pipeline.predict(x_test)
        y_pred_proba = pipeline.predict_proba(x_test)[:, 1] if hasattr(clf, "predict_proba") else None
        
        # Calculate the accuracy and other metrics for the test set
        test_accuracy = accuracy_score(y_test, y_pred)
        test_precision = precision_score(y_test, y_pred, average='binary')
        test_recall = recall_score(y_test, y_pred, average='binary')
        test_f1 = f1_score(y_test, y_pred, average='binary')
        test_roc_auc = roc_auc_score(y_test, y_pred_proba) if y_pred_proba is not None else 'N/A'

        # Print test set metrics after model fitting
        print(f"\nTest Set Accuracy: {test_accuracy:.2f}")
        print(f"Test Set Precision: {test_precision:.2f}")
        print(f"Test Set Recall: {test_recall:.2f}")
        print(f"Test Set F1-Score: {test_f1:.2f}")
        print(f"Test Set ROC-AUC: {test_roc_auc}")

        

            
        training_results[f'{clf_name}_{encoder_name}'] = [cv_results['test_accuracy'].mean(), cv_results['test_precision'].mean(), cv_results['test_recall'].mean(), cv_results['test_f1'].mean(),cv_results['test_roc_auc'].mean() ]
        testing_results[f'{clf_name}_{encoder_name}'] = [test_accuracy,   test_precision,   test_recall,   test_f1,   test_roc_auc ]
        
      
        
```
